# Final step for dataset prepartion

Here the outcome from "Dataset_Preparation" and "Dataset_Preparation_Last12" are subtracted and the difference is kept as final information needed for the classifying models training. 
The code loads 4 CSV files because the "Data_Collection" file code has been ran twice, with the intent to have bigger sample size. Becuase each file from the data collection process is split into overall performance and last 12 matches performance, we therefore have 4 files in this step below!

```python
df1 = pd.read_csv('TransformedDatasetSilver4.csv')
df2 = pd.read_csv('TransformedDatasetSilver4.2.csv')
df3= pd.read_csv('last_12_matches_part1.csv')
df4 = pd.read_csv('last_12_matches_part2.csv')

data = [df1,df2]
data2 = [df3,df4]

df = pd.concat(data, ignore_index=True, sort=False)
df2 = pd.concat(data2, ignore_index=True, sort=False)
pd.set_option('display.max_rows', None, 'display.max_columns', None)
```
