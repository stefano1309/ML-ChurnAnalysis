# Dataset Preparation for last 12 matches

This code is replicating the process of the file "Dataset_Preparation", but it includes only the last 12 matches a player has played. This is done becuse in the next step it will be calculated the difference between the output here against the output generate with the "Dataset_Preparation" file.

```python
df = pd.read_csv("Silver4_Dataset.csv")
```

```python
df
```

```python
#defines a function to calculate whether a player is a churner or not
def chrun_function(row):
    if row['gameCreation'] > 1656590400000:  #this is the time of 20 july - 20 days. WHere 20 july is data collection day
        val = 'False'
    else:
        val = 'True'
    return val
```

```python
#applies the function to the dataframe and then creates a column to store the result
df['Churner'] = df.apply(chrun_function, axis=1)
```

```python
df = df[df["puuid"].str.contains("#NAME?") == False]
df
```

```python
df = df.groupby('puuid').apply(lambda x: x.sort_values('gameCreation').tail(12)).reset_index(drop=True)
```

```python
df.head(150)
```

```python
df_avg = df.groupby(['puuid']).agg({'assists':[np.mean],'deaths':[np.mean],'kills':[np.mean],
                                          'bountyLevel':[np.mean],'champExperience':[np.mean],'champLevel':[np.mean],'damageDealtToBuildings':[np.mean],
                                          'damageDealtToObjectives':[np.mean],'damageDealtToTurrets':[np.mean],'damageSelfMitigated':[np.mean],'inhibitorTakedowns':[np.mean],
                                          'inhibitorsLost':[np.mean],'killingSprees':[np.mean],'pentaKills':[np.mean],'quadraKills':[np.mean],
                                          'timePlayed':[np.mean],'totalTimeSpentDead':[np.mean],'wardsKilled':[np.mean],'wardsPlaced':[np.mean],
                                          'baronKills':[np.mean],'acesBefore15Minutes':[np.mean],'controlWardsPlaced':[np.mean],'damagePerMinute':[np.mean],
                                          'goldPerMinute':[np.mean],'killParticipation':[np.mean],'laneMinionsFirst10Minutes':[np.mean],'maxCsAdvantageOnLaneOpponent':[np.mean],
                                          'skillshotsHit':[np.mean],'damageTakenOnTeamPercentage':[np.mean],'gameLength':[np.mean]})
```

```python
df_avg
```

```python
df_boolean_variables = df.drop(columns=['assists','deaths','kills','bountyLevel','champExperience','champLevel','damageDealtToBuildings','damageDealtToObjectives','damageDealtToTurrets','damageSelfMitigated','inhibitorsLost','killingSprees',
                                       'pentaKills','quadraKills','timePlayed','totalTimeSpentDead','wardsKilled','wardsPlaced','baronKills','deaths','acesBefore15Minutes','controlWardsPlaced','damagePerMinute','goldPerMinute',
                                       'killParticipation','laneMinionsFirst10Minutes','maxCsAdvantageOnLaneOpponent','skillshotsHit','damageTakenOnTeamPercentage','kills','gameLength','gameCreation','id','accountId','profileIconId',
                                        'leaguePoints','revisionDate','summonerId','summonerName','tier','rank','leagueId','queueType',
                                        'tripleKillstripleKills','detectorWardsPlaced','doubleKills','dragonKills','inhibitorTakedowns',
                                       'summonerLevel_x','veteran','inactive','individualPosition','lane'])
df_boolean_variables
```

```python
df_boolean_percentages = df_boolean_variables.groupby('puuid')['firstBloodAssist','firstBloodKill','firstTowerAssist','firstTowerKill',
                                            'gameEndedInEarlySurrender','gameEndedInSurrender','teamEarlySurrendered','win',
                                            'freshBlood','hotStreak'].apply(lambda x: np.sum(x)/len(x)*100).round(2)
df_boolean_percentages
```

```python
df_boolean_variables['Churner'] = (df_boolean_variables['Churner'] == 'True').astype(int)
```

```python
df_boolean_variables.head()
```

```python
churned_column = df_boolean_variables.groupby(['puuid']).agg({'Churner':[np.mean]})
churned_column
```

```python
churned_column.columns = churned_column.columns.droplevel(1) 
#df_boolean_percentages.columns = df_boolean_percentages.columns.droplevel(1) 
df_avg.columns = df_avg.columns.droplevel(1) 

```

```python
#defines a function to calculate whether a player is a churner or not
def churn_avg_transformation(row):
    if row['Churner'] < 1:  #this is the time of 8 july - 21 days. WHere 8 july is data collection day
        val = 'False'
    else:
        val = 'True'
    return val
```

```python
churned_column_df = churned_column
average_numbers_df = df_avg
percentages_booleans_df = df_boolean_percentages
```

```python
#applies the function to the dataframe and then creates a column to store the result
churned_column_df['Churner'] = churned_column_df.apply(churn_avg_transformation, axis=1)
```

```python
finaldDataset = average_numbers_df.merge(percentages_booleans_df,on='puuid').merge(churned_column_df,on='puuid')
```
