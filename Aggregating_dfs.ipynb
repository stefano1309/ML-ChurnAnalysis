# Merging all the CSV files obtained by running the code in "Data_Collection"

The code checks for all CSV files in the give directory and merge them togheter into a single dataframe

```python
files = glob.glob("DatasetThesisSilver/*/*.csv")
dfs = [pd.read_csv(f) for f in files]

dataset = pd.concat(dfs,ignore_index=True)
```
