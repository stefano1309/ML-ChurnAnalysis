#  Get match details

For each "gameID code" the API is requested to provide addtional and specific match details. The API response is a nested dictionary and only the necessay / interesting features are copied. The value of "x" (0,62,123) in the while loop is manually adjusted at different iterations, as well as "matchidtest" (ID1 TO ID100). The API crashes frequently, there are errors and/or downtime maintanace/heavy traffic, therefore the matches are passed in batches and later on will be merged in the other jupyter notebook file.

```python
matchidtest = list_of_match_numbers['ID100'] #edit for changing columns
x=62    #0,62,123 Values of X have to be changed until all matches are examinated

participants = []
while x < 123:        #62,123,185 values need to be changed incrementally, as the code can't handle all requests at once
    for _ in list_of_match_numbers.iterrows():
        matchIDtesting = matchidtest.iloc[x]
        matchdetailsapitest = watcher.match.by_id(my_region,matchIDtesting)
        
        for row in matchdetailsapitest['info']['participants']:
            try:
                participants_row = {}
                participants_row['gameCreation'] = matchdetailsapitest['info']['gameCreation']
                participants_row['puuid'] = row['puuid']
                participants_row['assists'] = row['assists']
                participants_row['bountyLevel'] = row['bountyLevel']
                participants_row['champExperience'] = row['champExperience']
                participants_row['champLevel'] = row['champLevel']
                participants_row['damageDealtToBuildings'] = row['damageDealtToBuildings']
                participants_row['damageDealtToObjectives'] = row['damageDealtToObjectives']
                participants_row['damageDealtToTurrets'] = row['damageDealtToTurrets']
                participants_row['damageSelfMitigated'] = row['damageSelfMitigated']
                participants_row['deaths'] = row['deaths']
                participants_row['detectorWardsPlaced'] = row['detectorWardsPlaced']
                participants_row['doubleKills'] = row['doubleKills']
                participants_row['dragonKills'] = row['dragonKills']
                participants_row['firstBloodAssist'] = row['firstBloodAssist']
                participants_row['firstBloodKill'] = row['firstBloodKill']
                participants_row['firstTowerAssist'] = row['firstTowerAssist']
                participants_row['firstTowerKill'] = row['firstTowerKill']
                participants_row['gameEndedInEarlySurrender'] = row['gameEndedInEarlySurrender']
                participants_row['gameEndedInSurrender'] = row['gameEndedInSurrender']
                participants_row['individualPosition'] = row['individualPosition']
                participants_row['inhibitorTakedowns'] = row['inhibitorTakedowns']
                participants_row['inhibitorsLost'] = row['inhibitorsLost']
                participants_row['killingSprees'] = row['killingSprees']
                participants_row['kills'] = row['kills']
                participants_row['lane'] = row['lane']
                participants_row['largestKillingSpree'] = row['largestKillingSpree']
                participants_row['largestMultiKill'] = row['largestMultiKill']
                participants_row['longestTimeSpentLiving'] = row['longestTimeSpentLiving']
                participants_row['objectivesStolen'] = row['objectivesStolen']
                participants_row['objectivesStolenAssists'] = row['objectivesStolenAssists']
                participants_row['pentaKills'] = row['pentaKills']
                participants_row['quadraKills'] = row['quadraKills']
                participants_row['summonerLevel'] = row['summonerLevel']
                participants_row['teamEarlySurrendered'] = row['teamEarlySurrendered']
                participants_row['timePlayed'] = row['timePlayed']
                participants_row['totalTimeSpentDead'] = row['totalTimeSpentDead']
                participants_row['tripleKillstripleKills'] = row['tripleKills']
                participants_row['timePlayed'] = row['timePlayed']
                participants_row['wardsKilled'] = row['wardsKilled']
                participants_row['wardsPlaced'] = row['wardsPlaced']
                participants_row['win'] = row['win']
                participants_row['baronKills'] = row['baronKills']
                participants_row['acesBefore15Minutes'] = row['challenges']['acesBefore15Minutes']
                participants_row['controlWardsPlaced'] = row['challenges']['controlWardsPlaced']
                participants_row['damagePerMinute'] = row['challenges']['damagePerMinute']
                participants_row['damageTakenOnTeamPercentage'] = row['challenges']['damageTakenOnTeamPercentage']    
                participants_row['gameLength'] = row['challenges']['gameLength']
                participants_row['goldPerMinute'] = row['challenges']['goldPerMinute']
                participants_row['killParticipation'] = row['challenges']['killParticipation']
                participants_row['laneMinionsFirst10Minutes'] = row['challenges']['laneMinionsFirst10Minutes']
                participants_row['maxCsAdvantageOnLaneOpponent'] = row['challenges']['maxCsAdvantageOnLaneOpponent']
                participants_row['skillshotsHit'] = row['challenges']['skillshotsHit']        
                participants.append(participants_row)
                
            except:
                continue
            
           
        df = pd.DataFrame(participants)
            
        x=x+1
        break
```
