# Create a function to identify churners

Based on the gameCreation variable, the funtion establishes whether the game has occured at least 20 days prior the day data was collected from the API. In next sections, the code will decide if a player is a chuner by a simple rule. If in the player's matches there is at least one game that has occured within the 20 days range, then this player is not a churner. On the other hand, if all matches are older than 20 days, then the player is classified as churner.

```python
#defines a function to calculate whether a player is a churner or not
def chrun_function(row):
    if row['gameCreation'] > 1656590400000:  #this is the time of 20 july - 20 days. WHere 20 july is data collection day
        val = 'False'
    else:
        val = 'True'
    return val
```

```python
#applies the function to the dataframe and then creates a column to store the result
df['Churner'] = df.apply(chrun_function, axis=1)
```

# groupped by puuid and boolean calculated as percentage

For boolean features, this code groups players' matches and transform the boolean variable into a percentage

```python
df_boolean_percentages = df_boolean_variables.groupby('puuid')['firstBloodAssist','firstBloodKill','firstTowerAssist','firstTowerKill',
                                            'gameEndedInEarlySurrender','gameEndedInSurrender','teamEarlySurrendered','win',
                                            'freshBlood','hotStreak'].apply(lambda x: np.sum(x)/len(x)*100).round(2)
df_boolean_percentages
```

# Column churn entries transformation into single row

From the previous code section which assigne a churn / not churn label based on the created function, some players may have some of the matches flagged as churner and some of them flagged as not churner. To summarize the information and to finally decide whether the player is a churner or not, the values are transformed into binary 0/1 values. A value of 0 indicates the match is labelled as "false", meaning "not churner" and if the value is 1 it means the match was labelled "true", meaning "churner". By calculating the mean of 0/1 values it is possible to state that if the mean equals to 1 than the players matches must all be 1 (True/churner) and if the mean is less than 1 it means there is at least one match with a value 0, meaning at least one match was played within the 20 days and therefore this player is not a churner.

```python
df_boolean_variables['Churner'] = (df_boolean_variables['Churner'] == 'True').astype(int)
```

```python
churned_column = df_boolean_variables.groupby(['puuid']).agg({'Churner':[np.mean]})
churned_column
```

```python
churned_column.columns = churned_column.columns.droplevel(1) 
#df_boolean_percentages.columns = df_boolean_percentages.columns.droplevel(1) 
df_avg.columns = df_avg.columns.droplevel(1) 
```

```python
churned_column_df = churned_column
average_numbers_df = df_avg
percentages_booleans_df = df_boolean_percentages
```

```python
#defines a function to calculate whether a player is a churner or not
def churn_avg_transformation(row):
    if row['Churner'] < 1: 
        val = 'False'
    else:
        val = 'True'
    return val
```

```python
#applies the function to the dataframe and then creates a column to store the result
churned_column_df['Churner'] = churned_column_df.apply(churn_avg_transformation, axis=1)
```

# Merge the three df

This code merges the three dataframes created earlier. One being the average of the numerical features, one being the percentage transformation of the boolean features and the last one being the labelling of player being churners or not. The 3 dataframes are merged on the column puuid which identifies the players.

```python
finaldDataset = average_numbers_df.merge(percentages_booleans_df,on='puuid').merge(churned_column_df,on='puuid')
```
